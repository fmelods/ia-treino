```python
#Realizando todas as importações necesárias para realização da análise de dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Visualizando e lendo mais informações sobre o dataset
data = pd.read_csv('data.csv')
data.head()
```

```python
#Contando quantas linhas e colunas temos dentro do arquivo
linhas, colunas = data.shape

# Exibindo o resultado
print(f"O dataset possui {linhas} linhas e {colunas} colunas.")
```

```python
# Com o código abaixo, estou verificando se há dados faltantes nas colunas!
print(data.isnull().sum())
```

```python
#Com o código anterior identificamos 5 (cinco) dados faltantes dentro da coluna "Calories", faremos então a mediana para preenchimento dos dados!
data['Calories'] = data['Calories'].fillna(data['Calories'].median())
```

```python
#Como podemos ver, agora não há mais nenhum tipo de dado faltante, pois identificamos a problemática dentro de Calories e resolvemos com a mediana!
```

```python
print(data.describe())
```

```python
#Identifiquei um tipo de ralação muito interessante, entre Duração vs. Calorias
plt.figure(figsize=(8,6))
sns.regplot(x='Duration', y='Calories', data=data, scatter_kws={"s": 50}, line_kws={"color": "red"})
plt.title('Relação entre Duração do Exercício e Calorias Queimadas (com Linha de Tendência)')
plt.xlabel('Duração do Exercício (minutos)')
plt.ylabel('Calorias Queimadas')
plt.show()
```

```python
# Podemos também imaginar que alguém mais "leigo" com gráficos estatísticos fará a leitura e não entendeu muito bem, então farei a criação de um mais simples!
data['Duration_Cat'] = pd.cut(data['Duration'], bins=[0, 30, 60, 90, 120, 150, 200, 250, 300], labels=['0-30 min', '30-60 min', '60-90 min', '90-120 min', '120-150 min', '150-200 min', '200-250 min', '250-300 min'])

# Posso calcular a média de calorias para cada categorai de duração com a linha abaixo:
mean_calories = data.groupby('Duration_Cat')['Calories'].mean()

# Criando o gráfico de barras (um gráfico relativamente mais simples para visualizações)
plt.figure(figsize=(10,6))
mean_calories.plot(kind='bar', color='skyblue')
plt.title('Média de Calorias Queimadas por Intervalos de Duração de Exercício')
plt.xlabel('Duração do Exercício (em minutos)')
plt.ylabel('Média de Calorias Queimadas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Vou criar abaixo o terceiro exemplo de gráfico para ainda mais formas de visualizações, o Gráfico de linha para mostrar a tendência:
plt.figure(figsize=(10,6))
sns.lineplot(x='Duration', y='Calories', data=data, marker='o', color='green')
plt.title('Tendência de Calorias Queimadas conforme a Duração do Exercício')
plt.xlabel('Duração do Exercício (minutos)')
plt.ylabel('Calorias Queimadas')
plt.grid(True)
plt.show()
```

```python
# Calcularemos abaixo: a correlação entre 'Duration' e 'Calories' para verificar a relação linear entre essas variáveis:
correlation = data[['Duration', 'Calories']].corr()
# Sendo assim, visualizaremos a matriz de correlação, onde valores próximos de 1 indicam uma forte correlação positiva!
print("Correlação entre Duração e Calorias:\n", correlation)
```

```python
# Análise descritiva das variáveis numéricas e entender melhor as variáveis
# Média
mean_duration = data['Duration'].mean()
mean_pulse = data['Pulse'].mean()
mean_maxpulse = data['Maxpulse'].mean()
mean_calories = data['Calories'].mean()

# Mediana
median_duration = data['Duration'].median()
median_pulse = data['Pulse'].median()
median_maxpulse = data['Maxpulse'].median()
median_calories = data['Calories'].median()

# Máximo
max_duration = data['Duration'].max()
max_pulse = data['Pulse'].max()
max_maxpulse = data['Maxpulse'].max()
max_calories = data['Calories'].max()

# Mínimo
min_duration = data['Duration'].min()
min_pulse = data['Pulse'].min()
min_maxpulse = data['Maxpulse'].min()
min_calories = data['Calories'].min()

# Desvio padrão
std_duration = data['Duration'].std()
std_pulse = data['Pulse'].std()
std_maxpulse = data['Maxpulse'].std()
std_calories = data['Calories'].std()

# Exibindo todos os resultados acima para melhores visualizações
print("Médias:")
print(f"Duração: {mean_duration:.2f}, Pulso: {mean_pulse:.2f}, Maxpulso: {mean_maxpulse:.2f}, Calorias: {mean_calories:.2f}\n")

print("Medianas:")
print(f"Duração: {median_duration}, Pulso: {median_pulse}, Maxpulso: {median_maxpulse}, Calorias: {median_calories}\n")

print("Máximos:")
print(f"Duração: {max_duration}, Pulso: {max_pulse}, Maxpulso: {max_maxpulse}, Calorias: {max_calories}\n")

print("Mínimos:")
print(f"Duração: {min_duration}, Pulso: {min_pulse}, Maxpulso: {min_maxpulse}, Calorias: {min_calories}\n")

print("Desvios Padrão:")
print(f"Duração: {std_duration:.2f}, Pulso: {std_pulse:.2f}, Maxpulso: {std_maxpulse:.2f}, Calorias: {std_calories:.2f}\n")
```
